# Видаляємо усі рядки у вихідній таблиці за допомогою методу dropna

```python
df = df.dropna()
df.head(5)
```

# Визначаємо новий розмір таблиці

# Створюємо нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = df.loc[df['Мова.програмування'] == 'Python'].copy()
python_data.head()
```

# Визначаємо розмір таблиці python_data

```python
print(f"Кількість рядків та стовпців: {python_data.shape}")
```

# Виконаємо групування за стовпчиком "Посада"

```python
grouped = python_data.groupby('Посада')
grouped.groups
```

# Створюємо новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконуємо агрегацію даних та знаходимо мінімальне та максимальне значення у стовпчику "Зарплата.на.місяць"

```python
salary_agg = python_data.groupby('Посада')['Зарплата.в.місяць'].agg(['min', 'max'])
salary_agg.head()
```

# Створюємо функцію fill_avg_salary, яка повертає середнє значення заробітної плати на місяць

```python
def fill_avg_salary(x):
    return x.mean()
```

# Використовуємо fill_avg_salary для методу apply та створюємо новий стовпчик "avg"

```python
python_data['avg'] = python_data[['Зарплата.в.місяць']].apply(fill_avg_salary, axis=1)

python_data.head()
```

# Створюємо описову статистику для нового стовпчика

```python
python_data['avg'].describe().apply(lambda x: format(x, '.2f'))
```

# Зберігаємо отриману таблицю в CSV файл

```python
python_data.to_csv('python_data.csv', index=False)
```
